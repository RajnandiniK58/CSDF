```python
from captcha.image import ImageCaptcha
import random
import string
from IPython.display import Image as IPImage, display

# Function to generate random captcha text
def generate_captcha_text(length=6):
    chars = string.ascii_letters + string.digits
    return ''.join(random.choice(chars) for _ in range(length))

# Function to create captcha image
def generate_captcha_image(captcha_text, file_name="captcha.png"):
    image = ImageCaptcha(width=280, height=90)
    image.write(captcha_text, file_name)
    print(f"CAPTCHA saved as {file_name}")
    display(IPImage(filename=file_name))  # Show inside Jupyter

# Function to verify captcha
def verify_captcha(original, user_input):
    return original == user_input

# Main execution
captcha_text = generate_captcha_text(6)
captcha_text

# Save and display image
generate_captcha_image(captcha_text)
user_input = input("Enter the CAPTCHA from the image: ")

if verify_captcha(captcha_text, user_input):
    print("CAPTCHA Matched ✅")
else:
    print("CAPTCHA Not Matched ❌")
```
